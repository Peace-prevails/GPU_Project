```python
import pandas as pd
import statsmodels.api as sm
from scipy import stats
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_excel("data.xlsx")
df
```

```python
df_standardized = df
columns_to_standardize = ['N', 'FLOPs', 'Memory']
scaler = StandardScaler()
df_standardized[columns_to_standardize] = scaler.fit_transform(df[columns_to_standardize])
df_standardized
```

```python
features = ['N', 'FLOPs', 'Overhead', 'isStrongDependent','Memory']
label = ['SpeedUp']
X = df_standardized[features]
Y = df_standardized[label]
X_new = sm.add_constant(X)
```

```python
test_index1 = 0
test_index2 = 12
X_train = X_new.drop([test_index1, test_index2])
X_test = X_new.iloc[[test_index1, test_index2]]
Y_train = Y.drop([test_index1, test_index2])
Y_test = Y.iloc[[test_index1, test_index2]]
```

```python
regcorr = sm.OLS(Y_train, X_train)
stats = regcorr.fit()
print(stats.summary())
```

```python
stats.predict(X_test)
```

```python
Y_test
```

```python
model = sm.GLM.from_formula("SingleOrDual ~ N + FLOPs + Overhead + isStrongDependent + Memory", family = sm.families.Binomial(), data = df_standardized.drop([test_index1,test_index2]))
res = model.fit()
res.summary()
```

```python
res.predict(X_test)
```
